```python
# preprocessing the data

from sklearn import preprocessing

# Iterating over parameters

best_score = 0

for i in range(1,6):
    if i==1:
        scale = preprocessing.Normalizer().fit(X_train)
    elif i==2:
        scale = preprocessing.MaxAbsScaler().fit(X_train)
    elif i==3:
        scale = preprocessing.MinMaxScaler().fit(X_train)
    elif i==4:
        scale = preprocessing.KernelCenterer().fit(X_train)
    else:
        scale = preprocessing.StandardScaler().fit(X_train)
    X_trainS = scale.transform(X_train)
    X_testS = scale.transform(X_test)
    for C in Cs:
        for gamma in gammas:
            svc = SVC(C = C,gamma=gamma)
            svc.fit(X_trainS,y_train)
            score = svc.score(X_testS, y_test)
            if score > best_score:
                print("#############################")
                print("gamma = ",gamma)
                print("C = ",C)
                print("i = ",i)
                print("score = ",score)
                best_score = score
```

```python
# preprocessing the data

from sklearn.decomposition import PCA

# Iterating over parameters

best_score = 0

for i in range(5,6):
    if i==1:
        scale = preprocessing.Normalizer().fit(X_train)
    elif i==2:
        scale = preprocessing.MaxAbsScaler().fit(X_train)
    elif i==3:
        scale = preprocessing.MinMaxScaler().fit(X_train)
    elif i==4:
        scale = preprocessing.KernelCenterer().fit(X_train)
    else:
        scale = preprocessing.StandardScaler().fit(X_train)
    X_trainS = scale.transform(X_train)
    X_testS = scale.transform(X_test)
    for j in range(4,15):
        pca = PCA(n_components=j, svd_solver="full")
        pca_scale = pca.fit(X_trainS)
        X_trainSPCA = pca_scale.transform(X_trainS)
        X_testSPCA = pca_scale.transform(X_testS)
        for C in Cs:
            for gamma in gammas:
                svc = SVC(C = C,gamma=gamma)
                svc.fit(X_trainSPCA,y_train)
                score = svc.score(X_testSPCA, y_test)
                if score > best_score:
                    print("#############################")
                    print("gamma = ",gamma)
                    print("C = ",C)
                    print("i = ",i)
                    print("j = ",j)
                    print("score = ",score)
                    best_score = score
```

```python
# preprocessing the data

from sklearn import manifold

# Iterating over parameters

best_score = 0

for i in range(5,6):
    if i==1:
        scale = preprocessing.Normalizer().fit(X_train)
    elif i==2:
        scale = preprocessing.MaxAbsScaler().fit(X_train)
    elif i==3:
        scale = preprocessing.MinMaxScaler().fit(X_train)
    elif i==4:
        scale = preprocessing.KernelCenterer().fit(X_train)
    else:
        scale = preprocessing.StandardScaler().fit(X_train)
    X_trainS = scale.transform(X_train)
    X_testS = scale.transform(X_test)
    for j in range(2,6):
        for k in range(4,7):
            iso = manifold.Isomap(n_neighbors=j, n_components=k)
            iso_scale = iso.fit(X_trainS)
            X_trainSiso = iso_scale.transform(X_trainS)
            X_testSiso = iso_scale.transform(X_testS)
            for C in Cs:
                for gamma in gammas:
                    svc = SVC(C = C,gamma=gamma)
                    svc.fit(X_trainSiso,y_train)
                    score = svc.score(X_testSiso, y_test)
                    if score > best_score:
                        print("#############################")
                        print("gamma = ",gamma)
                        print("C = ",C)
                        print("i = ",i)
                        print("j = ",j)
                        print("k = ",k)
                        print("score = ",score)
                        best_score = score
```
